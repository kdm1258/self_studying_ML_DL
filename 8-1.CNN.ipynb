# CNN(Convolutional Neural Network)
CNN(Convolutional Neural Network)은 주로 이미지나 영상 처리에 사용되는 인공 신경망의 일종이다. CNN은 이미지 데이터를 처리할 때 효과적인 방법으로, 입력 데이터를 여러 층을 거쳐 점차적으로 중요한 특징을 추출하여 분류나 예측을 수행한다.

1. 합성곱 층 (Convolutional Layer)
합성곱 층은 CNN의 핵심이다. 이 층에서는 필터(또는 커널)라고 불리는 작은 크기의 행렬을 사용하여 입력 데이터와 합성곱 연산을 수행한다. 이 연산을 통해 이미지를 작은 영역별로 분석하여, 이미지에서 중요한 특징을 추출한다. 예를 들어, 가장자리, 텍스처, 색상 등을 추출하는 데 유용하다.

2. 풀링 층 (Pooling Layer)
풀링 층은 이미지의 크기를 줄이면서, 중요한 정보를 유지하도록 돕는다. 주로 최대 풀링(max pooling)과 평균 풀링(average pooling) 방식이 사용된다. 풀링은 특성 맵(feature map)의 크기를 줄여서 계산 비용을 줄이고, 모델의 과적합을 방지하는 데 도움을 준다.

3. 완전 연결 층 (Fully Connected Layer)
합성곱과 풀링을 거친 후, 마지막에는 완전 연결 층이 있다. 이 층에서는 추출된 특징들을 바탕으로 최종적인 분류나 예측을 수행한다. 이미지의 특성을 모두 종합하여 결과를 도출하는 역할을 한다.

4. 활성화 함수 (Activation Function)
CNN에서는 각 층에서 비선형성을 추가하기 위해 활성화 함수가 사용된다. 대표적인 활성화 함수로는 **ReLU(Rectified Linear Unit)**가 있다. ReLU는 입력값이 0보다 크면 그대로 출력하고, 0 이하일 경우 0으로 출력하여 네트워크의 학습을 더욱 빠르고 효율적으로 만든다.

5. CNN의 특징
지역적인 연결: CNN은 각 뉴런이 입력 데이터 전체와 연결되는 것이 아니라, 지역적인 영역에만 연결된다. 이를 통해 모델이 더 효율적으로 학습할 수 있다.
파라미터 공유: 합성곱 연산에서 필터가 여러 위치에 걸쳐 동일하게 적용되므로, 파라미터 수가 크게 줄어들어 효율적인 학습이 가능하다.
계층적 특징 학습: CNN은 저수준의 특징(예: 모서리, 텍스처)부터 고수준의 특징(예: 얼굴, 객체)까지 계층적으로 학습할 수 있다.
6. CNN의 활용
CNN은 이미지 분류, 객체 탐지, 영상 분할, 얼굴 인식 등 다양한 분야에서 활용된다. 특히, 이미지 데이터를 다루는 많은 딥러닝 모델에서 CNN이 필수적으로 사용된다.

결론적으로, CNN은 이미지나 영상 데이터의 패턴을 인식하고 분석하는 데 매우 뛰어난 성능을 발휘하는 신경망 구조이다.

```python
from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3,3), activation = 'relu')
```

Conv2D의 첫 매개변수는 필터의 개수이다.
kernel_size는 커널의 크기를 설정하는 매개변수이다.

## 패딩과 스트라이드

컨볼루션 신경망(CNN)은 주로 이미지 처리에 사용되는 딥러닝 모델이다. 이 모델은 입력 이미지의 특징을 추출하기 위해 컨볼루션 층을 사용한다.

- 패딩(Padding): 입력 이미지의 가장자리에 빈 공간을 추가하여 입력 크기를 조정한다. 이를 통해 출력 크기를 원하는 대로 조절하거나, 가장자리 정보를 보존할 수 있다.

- 스트라이드(Stride): 필터가 이미지 위를 이동하는 간격을 나타낸다. 스트라이드가 클수록 출력 크기가 작아지고, 계산량이 줄어든다.

이 두 개념은 CNN의 성능과 효율성을 조절하는 데 중요한 역할을 한다.

### 패딩

- 세임 패딩(same padding) : 입력과 특성맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩하는 것을 세임 패딩이라고 함.

- 밸리드 패딩(valid padding) : 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우를 밸리드 패딩이라고 함.

위의 그림과 같이 패딩을 적용하면 모서리 픽셀이 합성곱에 참여하는 비율이 증가해 중앙부와 모서리 픽셀의 합성곱의 차가 줄어든다

```python
keras.layers.Conv2D(10, kernel_size = (3, 3), activation = 'relu', padding = 'same')
```

## 스트라이드(Stride)

스트라이드(Stride)는 컨볼루션 연산에서 필터가 입력 데이터 위를 이동하는 간격을 나타내는 것이다. 이는 CNN의 중요한 하이퍼파라미터 중 하나이며, 네트워크의 출력 크기와 계산 효율성에 영향을 준다.

스트라이드가 주는 영향
1. 출력 크기(Output Size):

  - 스트라이드가 1일 경우, 필터는 한 칸씩 이동하므로 출력 크기가 크게 줄지 않는다.

  - 스트라이드가 2 이상일 경우, 필터는 두 칸씩 이동하므로 출력 크기가 더욱 줄어드는 것이다.

2. 계산 효율성(Computational Efficiency):

  스트라이드가 클수록 필터의 이동 횟수가 줄어들어 계산량이 감소한다. 이는 네트워크의 속도와 효율성을 높이는 데 도움이 된다.

3. 정보 손실(Information Loss):

  스트라이드가 클수록 필터가 입력 데이터의 더 큰 부분을 건너뛰기 때문에 세밀한 특징을 놓칠 수 있다. 이는 출력 특징 맵의 해상도에 영향을 미칠 수 있다.


```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding = 'same', strides= 1)
```

## 풀링(Polling)

풀링은 주로 컨볼루션 신경망(CNN)에서 사용되는 연산이다. 풀링의 목적은 데이터의 공간 크기를 줄여 계산량을 감소시키고, 특징을 더욱 일반화하는 것이다.

**풀링의 종류:**
1. 최대 풀링(Max Pooling):

  풀링 영역에서 최대 값을 선택하는 것이다. 이는 가장 강한 특징을 유지하면서 공간 크기를 줄이는 데 효과적이다.

2. 평균 풀링(Average Pooling):

  풀링 영역에서 평균 값을 선택하는 것이다. 이는 전체적인 정보를 고르게 반영하면서 공간 크기를 줄이는 데 사용된다.

3. 글로벌 풀링(Global Pooling):

  전체 입력 데이터에서 하나의 값을 선택하는 것이다. 이는 주로 완전 연결 계층(Dense Layer)로 입력 데이터를 전달하기 전에 사용된다.

**풀링의 효과:**
1. 공간 크기 축소:

  풀링은 입력 데이터의 공간 크기를 줄이는 것이다. 이는 계산량을 줄이고, 네트워크의 학습 속도를 높이는 데 도움이 된다.

2. 특징 일반화:

  풀링은 특징을 더욱 일반화하여, 모델이 불필요한 세부 사항에 과적합(overfitting)되지 않도록 하는 것이다.

3. 정보 유지:

  최대 풀링의 경우, 가장 강한 특징을 유지하므로 중요한 정보를 보존하는 것이다. 평균 풀링의 경우, 전체적인 정보를 고르게 반영하여 균형 잡힌 특징을 유지하는 것이다.


풀링을 사용하는 이유는 스트라이드를 크게 하여 특성맵을 줄이는 것보다 풀링 층에서 크기를 줄이는 것이 경험적으로 더 나은 성능을 나타내기 때문이다.

```python
keras.layers.MaxPool2D(2)
```

```python
# 위의 코드와 같은 의미
keras.layers.MaxPool2D(2, strides=2, padding='valid')
```
